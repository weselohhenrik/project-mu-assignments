```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-1, 1, 0.1)
Y = np.arange(-1, 1, 0.1)
X, Y = np.meshgrid(X, Y)
Z = (X-Y)*(X - 1)*(X + 1)*(Y - 1)*(Y - 1)
#Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X, Y, Z, rstride = 10, cstride=5)
plt.show()
```
